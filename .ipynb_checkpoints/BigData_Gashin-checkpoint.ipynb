# Importing Libraries:

```python
import haversine
import pandas as pd
import math
import time
import numpy as np
import random
from datetime import datetime
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score,confusion_matrix
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score,confusion_matrix
from sklearn.model_selection import StratifiedKFold
from sklearn.ensemble import AdaBoostClassifier
from scipy import stats
```

# Loading the data and pre_processing it

```python
#Reading the Data
df= pd.read_csv("geolife_raw.csv", parse_dates=[1], infer_datetime_format=True)
#seperating Date and Time
df['collected_time']=pd.to_datetime(df['collected_time'])
temp = pd.DatetimeIndex(df['collected_time'])
df['Date'] = temp.date
df['hour'] = temp.hour
#removing class 'run' and 'motorcycle'
df=df[(df.transportation_mode !='run')].reset_index(drop=True)
df=df[(df.transportation_mode !='motorcycle')].reset_index(drop=True)
df=df.dropna(axis=0,how='any').reset_index(drop=True)

df.sort_values(['t_user_id','collected_time'],inplace=True)
```

# calculating the features:

```python
def Bearing(lat1,lon1,lat2,lon2):
    lat1 = math.radians(lat1)
    lat2 = math.radians(lat2)

    diffLong = math.radians(lon2 - lon1)

    x = math.sin(diffLong) * math.cos(lat2)
    y = math.cos(lat1) * math.sin(lat2) - (math.sin(lat1)
            * math.cos(lat2) * math.cos(diffLong))

    initial_bearing = math.atan2(x, y)
    initial_bearing = math.degrees(initial_bearing)
    b = (initial_bearing + 360) % 360
    return b
def Distance(lat1,lon1,lat2,lon2):
    p1=(lat1,lon1)
    p2=(lat2,lon2)
    h=haversine.haversine(p1,p2)*1000
    return h
def Time(t1,t2):
    t=(t2-t1).total_seconds()
    return t
    
```

```python
#Adding new columns from orginal ones with one shift to down to calculate the features
df['lat2']=df.latitude[1:]
df['lon2']=df.longitude[1:]
df['time2']=df.collected_time[1:]
df.lat2=df.lat2.shift(-1)
df.lon2=df.lon2.shift(-1)
df.time2=df.time2.shift(-1)
df.drop(df.index[-1], inplace=True)
#calculate the wanted features
start_time = time.time()
df['timediff']=df.apply(lambda x: Time(x['collected_time'],x['time2']), axis=1)
print("--- %s seconds ---" % (time.time() - start_time))
start_time = time.time()
df['bearing']=df.apply(lambda x: Bearing(x['latitude'], x['longitude'], x['lat2'],x['lon2']), axis=1)
print("--- %s seconds ---" % (time.time() - start_time))
start_time = time.time()
df['distance']=df.apply(lambda x: Distance(x['latitude'], x['longitude'], x['lat2'],x['lon2']), axis=1)
print("--- %s seconds ---" % (time.time() - start_time))
start_time = time.time()
df['speed']=df.distance/df.timediff
print("--- %s seconds ---" % (time.time() - start_time))
start_time = time.time()
df['acc']=df.speed/df.timediff
print("--- %s seconds ---" % (time.time() - start_time))
```

## Grouping the data:

```python
#slect wanted columns and grop by id and date
df= df[['t_user_id','Date','hour','transportation_mode','bearing','distance','speed','acc']].copy()
df=df.groupby(['t_user_id','Date','hour','transportation_mode'], as_index=False).apply(lambda x: x.iloc[:-1])
groupsA=df.groupby(['t_user_id','Date'],as_index=False)

#group by id and date and class then delete the groups with less than 10 rows
dataset=df.groupby(['t_user_id','Date','hour','transportation_mode'], as_index=False).filter(lambda x: len(x) > 10)
#group by id and date and class and calculate statistical measures
dataset=df.groupby(['t_user_id','Date','hour','transportation_mode'], as_index=False).agg([np.min,np.max,np.mean,np.median,np.std]).reset_index()
dataset.columns = ['ID','date','hour','t_mode',
                 'dis_min','dis_max','dis_mean','dis_median','dis_std',
                 'speed_min','speed_max','speed_mean','speed_median','speed_std',
                 'acc_min','acc_max','acc_mean','acc_median','acc_std',
                 'bearing_min','bearing_max','bearing_mean','bearing_median','bearing_std']
#delete unwanted columns and NAN rows
dataset.drop(['ID','date','hour'], axis=1,inplace=True)
dataset=dataset.dropna(axis=0,how='any').reset_index(drop=True)
dataset.to_csv('out3.csv')
```

## Prepraing the data for classification

```python

#shuffling the data
dataset_shuffle=dataset.sample(frac=1).reset_index(drop=True)
length=int(len(dataset_shuffle)*0.85)
#seperate train and test data
train_data = dataset_shuffle[:length]
test_data = dataset_shuffle[length:]
x_train=train_data[['dis_min','dis_max','dis_mean','dis_median','dis_std',
             'speed_min','speed_max','speed_mean','speed_median','speed_std',
             'acc_min','acc_max','acc_mean','acc_median','acc_std',
             'bearing_min','bearing_max','bearing_mean','bearing_median','bearing_std']]
y_train=train_data['t_mode'].tolist()

x_test=test_data[['dis_min','dis_max','dis_mean','dis_median','dis_std',
             'speed_min','speed_max','speed_mean','speed_median','speed_std',
             'acc_min','acc_max','acc_mean','acc_median','acc_std',
             'bearing_min','bearing_max','bearing_mean','bearing_median','bearing_std']]
y_test=test_data['t_mode'].tolist()
```

## Hierarchical Classifiers:

```python
def hierarchical_train_RF(x,x_test):
    #add columns for each layer to use
    x["l1"]=x['t_mode']
    x.loc[(x['t_mode']=='car') | (x['t_mode']=='bus') | (x['t_mode']=='train') | (x['t_mode']=='subway') | (x['t_mode']=='taxi'),'l1']="g1"
    x["l2"]=x["t_mode"]
    x.loc[(x['t_mode']=='car') | (x['t_mode']=='bus') | (x['t_mode']=='taxi'),'l2']="g2"
    x.loc[(x['t_mode']=='train') | (x['t_mode']=='subway'),'l2']="g3"
    
    #define classifiers fir each node
    clf_first=RandomForestClassifier(n_estimators=100)
    clf_second=RandomForestClassifier(n_estimators=100)
    clf_third1=RandomForestClassifier(n_estimators=100)
    clf_third2=RandomForestClassifier(n_estimators=100)
    
    
    
    #prepare data of each layer and fit the classifiers
    first_x=x.loc[:,['dis_min','dis_max','dis_mean','dis_median','dis_std',
             'speed_min','speed_max','speed_mean','speed_median','speed_std',
             'acc_min','acc_max','acc_mean','acc_median','acc_std',
             'bearing_min','bearing_max','bearing_mean','bearing_median','bearing_std']]
    first_y=x.loc[:,['l1']]   
    
    
    clf_first.fit(first_x,first_y.values.ravel())
    
    
    second_x= x.loc[(x['l1'] == 'g1'), ['dis_min','dis_max','dis_mean','dis_median',
                                           'dis_std','speed_min','speed_max','speed_mean',
                                           'speed_median','speed_std','acc_min','acc_max',
                                           'acc_mean','acc_median','acc_std','bearing_min',
                                           'bearing_max','bearing_mean','bearing_median',
                                           'bearing_std']]
    second_y= x.loc[(x['l1'] == 'g1'), ['l2']]
    
    clf_second.fit(second_x,second_y.values.ravel())
    
    third_1_x=x.loc[(x['l2'] == 'g2') , ['dis_min','dis_max','dis_mean','dis_median',
                                      'dis_std','speed_min','speed_max','speed_mean',
                                      'speed_median','speed_std','acc_min','acc_max',
                                      'acc_mean','acc_median','acc_std','bearing_min',
                                      'bearing_max','bearing_mean','bearing_median',
                                      'bearing_std']]
    third_1_y=x.loc[(x['l2'] == 'g2') , ['t_mode']]
    
    clf_third1.fit(third_1_x,third_1_y.values.ravel())
    
    third_2_x=x.loc[(x['l2'] == 'g3') , ['dis_min','dis_max','dis_mean','dis_median',
                                      'dis_std','speed_min','speed_max','speed_mean',
                                      'speed_median','speed_std','acc_min','acc_max',
                                      'acc_mean','acc_median','acc_std','bearing_min',
                                      'bearing_max','bearing_mean','bearing_median',
                                      'bearing_std']]
    third_2_y=x.loc[(x['l2'] == 'g3') , ['t_mode']]
    
    clf_third2.fit(third_2_x,third_2_y.values.ravel())
    #__________________________________________
    #test
    #prepare test data and use the classifier to predicts
    x_test=x_test[['dis_min','dis_max','dis_mean','dis_median','dis_std',
             'speed_min','speed_max','speed_mean','speed_median','speed_std',
             'acc_min','acc_max','acc_mean','acc_median','acc_std',
             'bearing_min','bearing_max','bearing_mean','bearing_median','bearing_std']].reset_index(drop=True)
    #first layer classifier
    out=clf_first.predict(x_test)
    out=pd.Series(out)
    # if predict to be g1 the send it to layer 2
    x_test2=x_test.loc[(out=='g1')]
    out2=clf_second.predict(x_test2)
    out.loc[(out=='g1')]=out2
    #send them to last layer classifiers according to predict of layer 2
    x_test3=x_test.loc[(out=='g2')]
    x_test4=x_test.loc[(out=='g3')]
    out3=clf_third1.predict(x_test3)
    out4=clf_third2.predict(x_test4)
    out.loc[(out=='g2')]=out3
    out.loc[(out=='g3')]=out4
    out=out.tolist()
    
    return out
    
    
#same hierarchical and codes just change the classifiers to AdaBoost

def hierarchical_train_AB(x,x_test):
    x["l1"]=x['t_mode']
    x.loc[(x['t_mode']=='car') | (x['t_mode']=='bus') | (x['t_mode']=='train') | (x['t_mode']=='subway') | (x['t_mode']=='taxi'),'l1']="g1"
    x["l2"]=x["t_mode"]
    x.loc[(x['t_mode']=='car') | (x['t_mode']=='bus') | (x['t_mode']=='taxi'),'l2']="g2"
    x.loc[(x['t_mode']=='train') | (x['t_mode']=='subway'),'l2']="g3"
    
    
    clf_first=AdaBoostClassifier(n_estimators=200)
    clf_second=AdaBoostClassifier(n_estimators=200)
    clf_third1=AdaBoostClassifier(n_estimators=200)
    clf_third2=AdaBoostClassifier(n_estimators=200)
    
    
    
    
    first_x=x.loc[:,['dis_min','dis_max','dis_mean','dis_median','dis_std',
             'speed_min','speed_max','speed_mean','speed_median','speed_std',
             'acc_min','acc_max','acc_mean','acc_median','acc_std',
             'bearing_min','bearing_max','bearing_mean','bearing_median','bearing_std']]
    first_y=x.loc[:,['l1']]   
    
    
    clf_first.fit(first_x,first_y.values.ravel())
    
    
    second_x= x.loc[(x['l1'] == 'g1'), ['dis_min','dis_max','dis_mean','dis_median',
                                           'dis_std','speed_min','speed_max','speed_mean',
                                           'speed_median','speed_std','acc_min','acc_max',
                                           'acc_mean','acc_median','acc_std','bearing_min',
                                           'bearing_max','bearing_mean','bearing_median',
                                           'bearing_std']]
    second_y= x.loc[(x['l1'] == 'g1'), ['l2']]
    
    clf_second.fit(second_x,second_y.values.ravel())
    
    third_1_x=x.loc[(x['l2'] == 'g2') , ['dis_min','dis_max','dis_mean','dis_median',
                                      'dis_std','speed_min','speed_max','speed_mean',
                                      'speed_median','speed_std','acc_min','acc_max',
                                      'acc_mean','acc_median','acc_std','bearing_min',
                                      'bearing_max','bearing_mean','bearing_median',
                                      'bearing_std']]
    third_1_y=x.loc[(x['l2'] == 'g2') , ['t_mode']]
    
    clf_third1.fit(third_1_x,third_1_y.values.ravel())
    
    third_2_x=x.loc[(x['l2'] == 'g3') , ['dis_min','dis_max','dis_mean','dis_median',
                                      'dis_std','speed_min','speed_max','speed_mean',
                                      'speed_median','speed_std','acc_min','acc_max',
                                      'acc_mean','acc_median','acc_std','bearing_min',
                                      'bearing_max','bearing_mean','bearing_median',
                                      'bearing_std']]
    third_2_y=x.loc[(x['l2'] == 'g3') , ['t_mode']]
    
    clf_third2.fit(third_2_x,third_2_y.values.ravel())
    #__________________________________________
    #test
    x_test=x_test[['dis_min','dis_max','dis_mean','dis_median','dis_std',
             'speed_min','speed_max','speed_mean','speed_median','speed_std',
             'acc_min','acc_max','acc_mean','acc_median','acc_std',
             'bearing_min','bearing_max','bearing_mean','bearing_median','bearing_std']].reset_index(drop=True)
    out=clf_first.predict(x_test)
    out=pd.Series(out)
    x_test2=x_test.loc[(out=='g1')]
    out2=clf_second.predict(x_test2)
    out.loc[(out=='g1')]=out2
    x_test3=x_test.loc[(out=='g2')]
    x_test4=x_test.loc[(out=='g3')]
    out3=clf_third1.predict(x_test3)
    out4=clf_third2.predict(x_test4)
    out.loc[(out=='g2')]=out3
    out.loc[(out=='g3')]=out4
    out=out.tolist()
    
    return out
    

    
```

## Classification

```python
pd.options.mode.chained_assignment = None
#train and test normal and Hierarchical clasiifiers
y1=hierarchical_train_RF(train_data,x_test)
y2=hierarchical_train_AB(train_data,x_test)
acc1=accuracy_score(y_test,y1)
acc2=accuracy_score(y_test,y2)
print("hierarchical randomforest's accuracy:",acc1)
print("hierarchical adaboost's accuracy:",acc2)

clf=RandomForestClassifier(n_estimators=100)
clf2=AdaBoostClassifier(n_estimators=200)
clf.fit(x_train,y_train)
clf2.fit(x_train,y_train)
y_t1=clf.predict(x_test)
y_t2=clf2.predict(x_test)
acc_n1=accuracy_score(y_test,y_t1)
acc_n2=accuracy_score(y_test,y_t2)
print("normal randomforest's accuracy:",acc_n1)
print("normal adaboost's accuracy:",acc_n2)
```

## Stratified 10_fold cross validation

```python

#use startifies cross validation to slice the data trian and test it 
#(10 fold cross validation) for Random forest noraml and hierarchical

x=dataset_shuffle[['dis_min','dis_max','dis_mean','dis_median','dis_std',
                 'speed_min','speed_max','speed_mean','speed_median','speed_std',
                 'acc_min','acc_max','acc_mean','acc_median','acc_std',
                 'bearing_min','bearing_max','bearing_mean','bearing_median','bearing_std']]

y=dataset_shuffle['t_mode']
score_hierarchical_RF=[]
score_RF=[]
skf = StratifiedKFold(n_splits=10)
# iterate over slices and add the accuracy to list of cross val
for train_index, test_index in skf.split(x, y):
    train_data =dataset_shuffle.loc[train_index,:]
    x_train=x.loc[train_index,:]
    y_train=dataset_shuffle.loc[train_index,'t_mode'].tolist()

    x_test=x.loc[test_index,:]
    y_test=dataset_shuffle.loc[test_index,'t_mode'].tolist()
    
    y=hierarchical_train_RF(train_data,x_test)
    acc1=accuracy_score(y_test,y)
    score_hierarchical_RF.append(acc1)
    
    clfn=RandomForestClassifier(n_estimators=100)
    clfn.fit(x_train,y_train)
    y_t=clfn.predict(x_test)
    acc2=accuracy_score(y_test,y_t)
    score_RF.append(acc2)

    
print('score_hierarchical_RF',score_hierarchical_RF)
print('score_RF',score_RF)
    
    
```

```python
# do the same Cross validation again for adaboost
x=dataset_shuffle[['dis_min','dis_max','dis_mean','dis_median','dis_std',
                 'speed_min','speed_max','speed_mean','speed_median','speed_std',
                 'acc_min','acc_max','acc_mean','acc_median','acc_std',
                 'bearing_min','bearing_max','bearing_mean','bearing_median','bearing_std']]

y=dataset_shuffle['t_mode']
score_hierarchical_AB=[]
score_AB=[]
skf = StratifiedKFold(n_splits=10)
for train_index, test_index in skf.split(x, y):
    train_data =dataset_shuffle.loc[train_index,:]
    x_train=x.loc[train_index,:]
    y_train=dataset_shuffle.loc[train_index,'t_mode'].tolist()

    x_test=x.loc[test_index,:]
    y_test=dataset_shuffle.loc[test_index,'t_mode'].tolist()
    
    y=hierarchical_train_AB(train_data,x_test)
    acc1=accuracy_score(y_test,y)
    score_hierarchical_AB.append(acc1)
    
    clfn=AdaBoostClassifier(n_estimators=200,algorithm="SAMME")
    clfn.fit(x_train,y_train)
    y_t=clfn.predict(x_test)
    acc2=accuracy_score(y_test,y_t)
    score_AB.append(acc2)

    
print('score_hierarchical_AB',score_hierarchical_AB)
print('score_AB',score_AB)
    
    
```

## Compare the classifiers with ttest and plots

```python
# compare the normal and hierarchical classifiers with ttest
t1,p1=stats.ttest_rel(score_RF,score_hierarchical_RF)
t2,p2=stats.ttest_rel(score_AB,score_hierarchical_AB)

if p1>0.05:
    print('null hypothesis is true for RandomForest and Hierarchical RanadomForest')
else:
    print('null hypothesis is false for RandomForest and Hierarchical RanadomForest')
if p2>0.05:
    print('null hypothesis is true for AdaBoost and Hierarchical AdaBoost')
else:
    print('null hypothesis is false for AdaBoost and Hierarchical AdaBoost')
    print(p1,p2)
```
